## Importing Libraries and Importing CSV File And Filling NAN VALUES in Dataset with 0
<h2> </h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"E:\income.csv")

# Fill missing values with 0
data = data.fillna(0)

data
```

## Making Income Groups
<h2> </h2>

```python

def income_category(income):
    if income < 80000:
        return 'Low Income'
    elif 80000 < income <= 120000:
        return 'Average Income'
    else:
        return 'High Income'


data['Income Group'] = data['Income'].apply(income_category)
data
```

## Sorting the data based on Name and Income columns Descending wise
<h2> </h2>

```python
data = data.sort_values(by=['Income', 'Name'], ascending=[False,True]) #This sort the data 

data

Top 5 Highest income persons
top_5 = data['Name'].head(5) #This will show top 5 highest income persons names
top_5_Income = data['Income'].head(5) #This will show top 5 highest income persons Income
top_5_Age = data['Age'].head(5) #This will show top 5 highest income persons Age


# Top 5 Lowest income persons
data = data.sort_values(by=['Income', 'Name'], ascending=[True,False])


# Top 5 lowest income persons
low_5 = data['Name'].head(5)
low_5_Income = data['Income'].head(5)
low_5_Age = data['Age'].head(5)

```

## Visualizing Top 5 Highest Income Persons And Lowest
<h2> </h2>

```python
# Visualizing top 5 highest income persons age
sns.boxplot(x=top_5, y=top_5_Age, data=data)
plt.show()

# Visualizing top 5 highest income persons income
sns.boxplot(x=top_5, y=top_5_Income, data=data)
plt.xlabel("Names")
plt.title("Top 5")
plt.show()

# Visualizing top 5 lowest income persons age
sns.boxplot(x=low_5, y=low_5_Age, data=data)
plt.title("Top 5 Lowest Income Persons Age")
plt.show()

# Visualizing top 5 lowest income persons Income
sns.boxplot(x=low_5, y=low_5_Income, data=data)
plt.title("Top 5 Lowest Income Persons Age")
plt.show()


```

## Finding Average Income based on Age:

<h2> </h2>

```python
# Grouping by age and calculating mean income
grouped = data.groupby('Age')
mean_income = grouped.mean('Income')
mean_income
```

## Categorizing Age And Counting Occurrences of Age Group

<h2>  </h2>

```python
# Function to categorize age
def age_levels(age):
    if 19 < age <=30:
        return '20-30'
    elif 30 < age <=40:
        return "31-40"
    else:
        return "41-50"

# Adding 'Age Group' column based on age

data['Age Group'] = data['Age'].apply(age_levels)
data

# Counting age group occurrences
count_of_20_30 = data['Age Group'].value_counts().get('20-30',0)
count_of_31_40 = data['Age Group'].value_counts().get('31-40',0)
count_of_41_50 = data['Age Group'].value_counts().get('41-50',0)

print(count_of_20_30)
print(count_of_31_40)
print(count_of_41_50)
```

## Pie Chart And Bar Chart for Age Group Chart
<h2> </h2>

```python
# Pie chart for age group counts
age_levels = ['20-30', '31-40', '41-50']
a = [count_of_20_30, count_of_31_40, count_of_41_50]
plt.figure(figsize=(8, 6))
plt.pie(a, labels=age_levels)
plt.title("Age Group Counts")
plt.show()

# Bar chart for age group counts
plt.bar(age_levels, a, color="Yellow")
plt.xlabel("Age Groups")
plt.show()
```

## Countplot for age groups and income groups
<h2> </h2>

```python
sns.countplot(x="Age Group", hue='Income Group', data=data)
plt.show()
```

## Decisian Tree Machine Learning Predicting the Income based on Age
<h2> </h2>

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

# Load the dataset
data = pd.read_csv(r"E:\income.csv")

# Fill missing values with 0
data = data.fillna(0)

# Selecting 'Age' as the input feature and 'Income' as the target variable
inputs = data[['Age']]
target = data['Income']

# Initialize and train the Decision Tree regressor
model = DecisionTreeRegressor()
model.fit(inputs, target)

# Predict income based on age
prediction = model.predict([[40]]) #here i am adding the age myself to see what income it will be for this age as per my model
print("Predicted Income:", prediction)
```
